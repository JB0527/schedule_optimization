```python
# 널값빼고 계산할수있게 다시
import pandas as pd

file_path = '/content/drive/MyDrive/MILab/Pipe/'
df = pd.read_csv(file_path + 'after_filtering_data.csv')

df['제품번호 그룹'] = df['제품번호'].str.split('-').str[0] + '-' + df['제품번호'].str.split('-').str[1]
grouped = df.groupby('제품번호 그룹')

new_cols = ['제품그룹', '외경/폭(mm)', '두께(mm)', '등록 길이(m)',
            'step1 평균 작업 시간(분)', 'step1 표준편차 작업 시간(분)', 'step2 평균 작업 시간(분)', 'step2 표준편차 작업 시간(분)',
            'step3 평균 작업 시간(분)', 'step3 표준편차 작업 시간(분)', 'step4 평균 작업 시간(분)', 'step4 표준편차 작업 시간(분)',
            'step5 평균 작업 시간(분)', 'step5 표준편차 작업 시간(분)', 'step6 평균 작업 시간(분)', 'step6 표준편차 작업 시간(분)',
            '제품개수']

results = []

for name, group in grouped:
    width = group.iloc[:, 1].tolist()[0]
    thickness = group.iloc[:, 2].tolist()[0]
    length = group.iloc[:, 3].tolist()[0]

    step1 = group.iloc[:, 7].dropna().tolist()  # 널값 제거
    step2 = group.iloc[:, 10].dropna().tolist()  # 널값 제거
    step3 = group.iloc[:, 13].dropna().tolist()  # 널값 제거
    step4 = group.iloc[:, 16].dropna().tolist()  # 널값 제거
    step5 = group.iloc[:, 19].dropna().tolist()  # 널값 제거
    step6 = group.iloc[:, 22].dropna().tolist()  # 널값 제거

    avg1 = round(sum(step1) / len(step1), 1) if step1 else None
    avg2 = round(sum(step2) / len(step2), 1) if step2 else None
    avg3 = round(sum(step3) / len(step3), 1) if step3 else None
    avg4 = round(sum(step4) / len(step4), 1) if step4 else None
    avg5 = round(sum(step5) / len(step5), 1) if step5 else None
    avg6 = round(sum(step6) / len(step6), 1) if step6 else None

    std1 = round(pd.Series(step1).std(), 1) if len(step1) > 1 else None
    std2 = round(pd.Series(step2).std(), 1) if len(step2) > 1 else None
    std3 = round(pd.Series(step3).std(), 1) if len(step3) > 1 else None
    std4 = round(pd.Series(step4).std(), 1) if len(step4) > 1 else None
    std5 = round(pd.Series(step5).std(), 1) if len(step5) > 1 else None
    std6 = round(pd.Series(step6).std(), 1) if len(step6) > 1 else None

    infos = [name, width, thickness, length, avg1, std1, avg2, std2, avg3, std3, avg4, std4, avg5, std5, avg6, std6, len(step1)]
    results.append(infos)

# 결과를 데이터프레임으로 변환합니다.
results_df = pd.DataFrame(results, columns=new_cols)

# 엑셀 파일로 저장합니다.
results_df.to_excel(file_path + 'pipe_avg_241006.xlsx', index=False)
```

```python
# 최종 sigma 거르기 !!!!!!!!!!!!
import pandas as pd

# 작업 시간 및 필터 기준
cols = ['step1 작업 시간(분)', 'step2 작업 시간(분)', 'step3 작업 시간(분)', 'step4 작업 시간(분)', 'step6 작업 시간(분)']
avg_cols = ['step1 평균 작업 시간(분)', 'step2 평균 작업 시간(분)', 'step3 평균 작업 시간(분)', 'step4 평균 작업 시간(분)', 'step6 평균 작업 시간(분)']
std_cols = ['step1 표준편차 작업 시간(분)', 'step2 표준편차 작업 시간(분)', 'step3 표준편차 작업 시간(분)', 'step4 표준편차 작업 시간(분)', 'step6 표준편차 작업 시간(분)']

path = '/content/drive/MyDrive/MILab/Pipe/'

df = pd.read_csv(path + "after_filtering_data.csv")
avg_df = pd.read_excel(path + "pipe_avg_241006.xlsx")

# 제거할 인덱스를 저장할 리스트
indexes_to_drop = []

# 각 제품 번호에 대해 반복
for index, row in df.iterrows():
    # 제품번호에서 그룹 이름 추출
    name = "-".join(row["제품번호"].split("-")[:2])

    # 평균 작업 시간 데이터에서 해당 제품 그룹의 행 찾기
    matching_rows = avg_df[avg_df["제품그룹"] == name]

    if not matching_rows.empty:

      for i in range(len(cols)):
        # 평균 및 표준편차 가져오기
        mean = matching_rows[avg_cols[i]].values[0]
        std = matching_rows[std_cols[i]].values[0]
        print(name,mean,std)

        lower_bound = round(mean - 0.25 * std, 3)
        upper_bound = round(mean + 0.75 * std, 3)
        # print(lower_bound, upper_bound)

        # 각 스텝별로 필터링 조건 확인
        if pd.notna(row[cols[i]]):
          # print(row[cols[i]], lower_bound, upper_bound)

          # 조건에 맞지 않는 경우, 현재 인덱스를 None으로 설정
          if not (row[cols[i]] <= upper_bound and row[cols[i]] >= lower_bound):
            df.at[index, cols[i]] = None  # 해당 칼럼의 값을 None으로 설정

pd.Series(indexes_to_drop).to_csv('/content/drive/MyDrive/MILab/Pipe/drop.csv')

# 조건에 맞지 않는 행 제거
df.drop(indexes_to_drop, inplace=True)

# 결과 저장
new_file_path = '/content/drive/MyDrive/MILab/Pipe/filter_pipe_final_241007.xlsx'
df.to_excel(new_file_path, index=False)

print(f"Complete filtering and saved to {new_file_path}")
```
